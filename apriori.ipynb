```python
# Function to calculate distance using Haversine formula
from geopy.distance import geodesic
def calculate_distance(row):
    wwtp_location = (row['LAT_WWTP'], row['LON_WWTP'])
    water_body_location = (row['LAT_OUT'], row['LON_OUT'])
    distance = geodesic(wwtp_location, water_body_location).kilometers
    return distance

# Add a new column dist indicating 'DISTANCE_TO_WATER_BODY' to the DataFrame
df['dist'] = df.apply(calculate_distance, axis=1)
```

## Apriori Algorithm 

```python
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
from mlxtend.preprocessing import TransactionEncoder
```

```python
df.columns
```

```python
data = df
relevant_attributes = ['RIVER_DIS','POP_SERVED','distance_bins']
data = data[relevant_attributes]

transactions = []
for index, row in data.iterrows():
    transaction = []
    for attribute in relevant_attributes:
        transaction.append(str(row[attribute]))
    transactions.append(transaction)

encoder = TransactionEncoder()
transaction_array = encoder.fit(transactions).transform(transactions)
encoded_data = pd.DataFrame(transaction_array, columns=encoder.columns_)

min_support = 0.1 
frequent_itemsets = apriori(encoded_data, min_support=min_support, use_colnames=True)

rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.5)

print("Association Rules:")
for index, rule in rules.iterrows():
    antecedents = ", ".join(rule['antecedents'])
    consequents = ", ".join(rule['consequents'])
    support = rule['support']
    confidence = rule['confidence']
    lift = rule['lift']
    print(f"Rule: {antecedents} => {consequents}")
    print(f"Support: {support:.3f}, Confidence: {confidence:.3f}, Lift: {lift:.3f}")
    print()

```
